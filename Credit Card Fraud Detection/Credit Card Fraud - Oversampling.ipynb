3. Feature Selection

```
df = spark.createDataFrame(df)
```

```
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import RandomForestClassifier

featureCols = df.columns
featureCols.remove('Class')

# Assuming your DataFrame df and you have columns listed in 'features'
assembler = VectorAssembler(inputCols=featureCols, outputCol="features_vec")
df_vector = assembler.transform(df)

rf = RandomForestClassifier(labelCol="Class", featuresCol="features_vec")
model = rf.fit(df_vector)

# Extract feature importances
importances = model.featureImportances

# Sort the feature importances in descending order
importances_list = [importances[i] for i in range(len(featureCols))]
importances_zip = zip(featureCols, importances_list)
importances_sorted = sorted(importances_zip, key=lambda x: x[1], reverse=True)

# Print the top 10 features
top_features = importances_sorted[:10]
print("Top 10 Features:")
for feature in top_features:
    print(feature[0])

# Feature importances visualization
import pandas as pd
import matplotlib.pyplot as plt

feature_importances_df = pd.DataFrame(importances_sorted, columns=['feature', 'importance'])

plt.figure(figsize=(10, 5))
plt.title("Feature Importances")
plt.bar(feature_importances_df['feature'], feature_importances_df['importance'])
plt.xticks(rotation=90)
plt.show()
```

```
import seaborn as sns
#get correlations of each features in dataset
df = df.toPandas()
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,10))

#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```
X = df[['V17', 'V14', 'V12', 'V7', 'V10']] 
y = df['Class']
```

```
from sklearn.model_selection import train_test_split
#I want to withhold 30 % of the trainset to perform the tests
#A stratified split ensures that the class distribution in both the training and testing sets reflects the original dataset.
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, stratify=y)
```

5. Oversampling imbalanced dataset

```
from imblearn import over_sampling
from imblearn.over_sampling import SMOTE

# Instantiate SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to oversample the minority class
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

y_resampled.value_counts()
```

6. Assemble feature columns into a feature vector

```
train = pd.concat([X_resampled,y_resampled],axis=1)
test = pd.concat([X_test,y_test],axis=1)
```

```
train = spark.createDataFrame(train)
test = spark.createDataFrame(test)
```

```
featureCols = train.columns
featureCols.remove('Class')

va = VectorAssembler(inputCols = featureCols, outputCol='features')
```

```
train_va = va.transform(train)
train_va.limit(3).display()
```

```
train_va = train_va.withColumnRenamed("Class", "label")
```

7. Training Model

```
from pyspark.ml import Pipeline
from pyspark.ml.classification import LogisticRegression, RandomForestClassifier, GBTClassifier, LinearSVC
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Define models
lr = LogisticRegression(labelCol="label", featuresCol="features")
rf = RandomForestClassifier(labelCol="label", featuresCol="features")
svm = LinearSVC(labelCol="label", featuresCol="features")
gbt = GBTClassifier(labelCol="label", featuresCol="features")

# Define parameter grids for each model
lrParamGrid = ParamGridBuilder() \
    .addGrid(lr.regParam, [0.1, 0.01]) \
    .build()

rfParamGrid = ParamGridBuilder() \
    .addGrid(rf.numTrees, [10, 50]) \
    .build()

svmParamGrid = ParamGridBuilder() \
    .addGrid(svm.maxIter, [10, 100]) \
    .build()

gbtParamGrid = ParamGridBuilder() \
    .addGrid(gbt.maxIter, [10, 20]) \
    .addGrid(gbt.maxDepth, [2, 4]) \
    .build()

# Set up the CrossValidator for each model
lrCv = CrossValidator(estimator=lr, estimatorParamMaps=lrParamGrid, evaluator=BinaryClassificationEvaluator())
rfCv = CrossValidator(estimator=rf, estimatorParamMaps=rfParamGrid, evaluator=BinaryClassificationEvaluator())
svmCv = CrossValidator(estimator=svm, estimatorParamMaps=svmParamGrid, evaluator=BinaryClassificationEvaluator())
gbtCv = CrossValidator(estimator=gbt, estimatorParamMaps=gbtParamGrid, evaluator=BinaryClassificationEvaluator())

# Fit models using CrossValidator and find the best model
lrModel = lrCv.fit(train_va)
rfModel = rfCv.fit(train_va)
svmModel = svmCv.fit(train_va)
gbtModel = gbtCv.fit(train_va)
```

8. Model Selection

```
evaluator = BinaryClassificationEvaluator(labelCol="label")

print("Logistic Regression AUC: " + str(evaluator.evaluate(lrModel.transform(train_va))))
print("Random Forest AUC: " + str(evaluator.evaluate(rfModel.transform(train_va))))
print("SVM AUC: " + str(evaluator.evaluate(svmModel.transform(train_va))))
print("Gradient Boosting Trees AUC: " + str(evaluator.evaluate(gbtModel.transform(train_va))))
```
